```python
#Gautami Kadam
import pandas as pd
import numpy as np
import tensorflow as tf
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
#Gautami Kadam
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense,Dropout
```

```python
#Gautami Kadam
df = pd.read_csv('train.csv')
```

```python
#Gautami Kadam
df = df.dropna()
```

```python
#Gautami Kadam
X = df.drop('label',axis=1)
```

```python
#Gautami Kadam
y = df['label']
```

```python
#Gautami Kadam
#Vocabulary size
voc_size = 5000
```

```python
#One hot representation
```

```python
#Gautami Kadam
mes = X.copy()
```

```python
#Gautami Kadam
mes['title'][1]
```

```python
#as we droped nan values so need to reset indexes
mes.reset_index(inplace=True)
```

```python
nltk.download('stopwords')
```

```python
#Gautami Kadam
ps = PorterStemmer()
string = "At: What time?"
match = re.sub("[^a-zA-Z]"," ",string)
print(match)
match = match.lower()
print(match)
match = match.split()
print(match)
review = [ps.stem(word) for word in match if not word in stopwords.words('english')]
print(review)
review = ' '.join(review)
print(review)
```

```python
#use of stem
#the idea of stemming is a sort of normalizing method. Many variataions 
#of words carry the same meaning, other than when tense is involved
#the reason why we stem is to shorten the lookup, and normalize sentences
ex_words = ["python","pythoner","pythoning","pythoned","pythonly"]
for w in ex_words:
    print(ps.stem(w))
```

```python
corpus = []
for i in range(0,len(mes)):
    rev = re.sub('[^a-zA-Z]',' ',mes['title'][i])
    rev = rev.lower()
    rev = rev.split()
    
    rev = [ps.stem(word) for word in rev if not word in stopwords.words('english')]
    rev = ' '.join(rev)
    corpus.append(rev)
```

```python
#Gautami Kadam
corpus
```

```python
onehot_rep = [one_hot(words,voc_size)for words in corpus]
onehot_rep
```

```python
length = 20
embed_docs = pad_sequences(onehot_rep,padding='pre',maxlen=length)
print(embed_docs)
```

```python
#Gautami Kadam
embed_docs[0]
```

```python
embed_vector_feature = 40
model = Sequential()
model.add(Embedding(voc_size,embed_vector_feature,input_length=length))
#one LSTM uses 100 neurons
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
#Gautami Kadam
len(embed_docs),y.shape
```

```python
X_final = np.array(embed_docs)
y_final = np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size=0.33,random_state=42)
```

```python
#model training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
#Gautami Kadam
embed_vector_feature = 40
model = Sequential()
model.add(Embedding(voc_size,embed_vector_feature,input_length=length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
y_pred = (model.predict(X_test)>0.5).astype("int32")
```

```python
#Gautami Kadam
confusion_matrix(y_test,y_pred)
```

```python
#Gautami Kadam
accuracy_score(y_test,y_pred)
```
